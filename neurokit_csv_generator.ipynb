```python
import neurokit2 as nk
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ecg100 = nk.ecg_simulate(duration=10, sampling_rate=100, noise=0.1, heart_rate=100)
ecg_df = pd.DataFrame({"ECG_100": ecg100})
```

```python
nk.signal_plot(ecg100)
plt.show()
```

```python
ecg_df.to_csv("dados_ecg.csv", index=False)
```
